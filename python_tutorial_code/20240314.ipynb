# Looking at conditionals

```python
df = pd.DataFrame({'ChildId':['id1', 'id2', 'id3', 'id4', 'id5'],
                   'Age first contact':[6,12,11,1,19],
                   'Gender':['M','m', 'F', '', 'F' ],
                   'Birthday':['01/01/2002', '02/02/2003', pd.NA, '03/03/2023', '06/01/2012'],
                   'Age':[5,12,11,6,2],
                   'CP Plan?':['Y', 'n', 'N', 'No', 'yES'],})
df['Birthday'] = pd.to_datetime(df['Birthday'], dayfirst=True)
df
```

```python
# select based on conditions

condition = df['Age'] >= 10

print(condition) # series - an array of True/False

over10 = df[condition] # subset slice of full df, change this and it could change df!!
over10copy = df[condition].copy

print(over10copy)
```

```python
# new df slice with only rows where the age of first contact column is greater than 9

condition2 = df['Age first contact'] >= 9
print(condition2)

first_contact_age_over9 = df[condition2]

print(first_contact_age_over9)
```

```python
error_rows = df[df['Age first contact'] > df['Age']]
print(error_rows)

not_error_rows = df[~(df['Age first contact'] > df['Age'])]
print(not_error_rows)
```

```python
print(((1 > 2) & (2 == 2))) # false, because 1 is not greater than 2 and both need to be true
print(((1 > 2) | (2 == 2))) # True because at least one of the components is true,
                            # even though 1 is not greater than 2
print(((3 > 2) | (2 == 2))) # Still true, as at least one is true
print(((3 > 2) & (2 == 2))) # Only strue because both are true
```

```python
# In this cell, writet a logical test that fails an and condition,
# but passes an or condition, and print both

print(((1 > 2) & (2 < 3)))

print(((1 > 2) | (2 < 3)))
```

```python
isna_condition = df['Birthday'].isna()

print(df[isna_condition])

notna_condition = (df['Birthday'].notna()) & (df['Age'] > 9)
print(df[notna_condition])

str_condition = df['CP Plan?'].str.lower().str.contains('y')
print(df[str_condition])

check_age_list = [12, 11]
condition_11_12 = df['Age'].isin(check_age_list)
print(df[condition_11_12])
```

```python
condition = (df['CP Plan?'].str.lower().str.contains('n')) & (df['Gender'].str.lower() == 'm')
print(df[condition])
```

```python
condition2 = df['Age'].isin([6, 9]) & (df['CP Plan?'].str.lower().str.contains('y')) # remember multiple .str occurrences
print(df[condition2])
```

# the fizzbang exercise

popular in coding interviews ten years ago

```python
# fizzbang
# take all numbers from 1-100
# if divisible by 3 print fizz
# if divisible by 5 print bang
# if divisible by 3 & 5 print fizzbang

numbers = list(range(1, 101))
print(numbers)

for i in numbers:
    p = ''
    if (i % 3 == 0):
        p = 'fizz'
    if (i % 5 == 0):
        p = p + "bang"
    if p == '':
        print(i)
    else:
        print(p)
```
